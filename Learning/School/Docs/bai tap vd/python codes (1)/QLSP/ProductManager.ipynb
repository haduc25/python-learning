```python
class Product:
    def __init__(self, id, name, price):
        self._id = id
        self._name = name
        self._price = price
    
    def getId(self):
        return self._id
    
    def setId(id):
        self._Id = id

    def getName(self):
        return self._name
    
    def setName(name):
        self._name = name

    def getPrice(self):
        return self._price
    
    def setName(price):
        self._price = price
    
    def show(self):
        print("{0}, {1}, {2}".format(self._id, self._name, self._price))
        

class ProductManager:
    _listProduct = []
    
    def generateId(self):
        maxId = 1

        if len(self._listProduct) > 0:
            maxId = self._listProduct[0].getId()

            for x in self._listProduct:
                if maxId < x.getId():
                    maxId = x.getId()

            maxId += 1
        return maxId 
    
    # Thêm một một product vào danh sách
    def input(self):
        id = self.generateId()
        name = input("Tên sản phẩm:")
        price = float(input("Giá: "))
        
        product  = Product(id, name, price)
        
        self._listProduct.append(product)
    
    def showAllProduct(self):
        for x in self._listProduct:
            x.show()
    
    def findById(self, id):
        product = None
        for x in self._listProduct:
            if x.getId() == id:
                product = x
                break
        
        return product
    
    def findByName(keyword):
        listProduct = []
        for x in self._listProduct:
            if x.upper() == keyword.upper():
                listProduct.append(x)                
        
        return listProduct    
    
    def deleteById(id):
        for x in self._listProduct:
            if x.getId() == id:
                self._listProduct.remove(x)
                return True
                
        return False
    
    def update(id):
        
        product = self.findById(id)
        
        if product == None:
            return False
        
        name = input("Tên sản phẩm:")
        price = float(input("Giá: "))
        
        product.setName(name);
        product.setPrice(price);
            
        return True
    
    def sortById(self):
        self._listProduct.sort(key = lambda x: x.getId(), reverse = False)
    
    def sortByName(self):
        self._listProduct.sort(key = lambda x: x.getName(), reverse = False)        
    
    def getAllProduct(self):
        return _listProduct
```
